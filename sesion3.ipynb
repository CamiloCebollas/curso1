## 1. NumPy

## 1.1 Concatenación y división de arreglos

La concatenación, o unión de dos arreglos en NumPy, se realiza principalmente a través de la función `concatenate`:

```
import numpy as np

x = np.array([1, 2, 3])

y = np.array([4, 5, 6])

np.concatenate([x, y])
```

Podemos concatenar más de dos arreglos al tiempo:

```
z = np.array([7, 8, 9])

np.concatenate([x, y, z])
```

`concatenate` también se puede usar para arreglos bidimensionales:

<p><img alt="Colaboratory logo" height="300px" src="https://i.imgur.com/KYPgvhf.png" align="left" hspace="10px" vspace="0px"></p>

Recuerde que las filas corresponden al eje 1 (axis = 1) y las columnas al eje 0 (axis = 0)


```
a = np.array([[8, 9], [10, 11], [12, 13]])
a
```

```
np.concatenate([a, a])

#np.concatenate([a, a], axis = 0)
```

Si queremos realizar la concatenación a lo largo del eje 1, debemos especificar el eje por medio de un argumento por palabra clave:

```
np.concatenate([a, a], axis=1)
```

La función `vstack()` es equivalente a la concatenación a lo largo del primer eje (axis=0). Al igual que `concatenate()`, recibe como argumento una secuencia de arreglos que, en este caso, deben tener la misma forma a lo largo de todos los ejes, excepto el primero. Los arreglos unidimensionales deben tener la misma longitud:

```
# dimensiones mixtas
a = np.array([1, 2, 3])

b = np.array([[3, 5, 6],
              [7, 8, 9]])

print(a.shape, b.shape)
np.vstack([a, b])
```

```
a = np.array([[1], 
              [2]])

b = np.array([[4],
              [5], 
              [6]])

print(a.shape, b.shape)
np.vstack([a,b])
```

```
#arreglos unidimensionales
a = np.array([1,2,3])

b = np.array([4,5,6,7])

print(a.shape, b.shape)
np.vstack([a, b])
```

Similarmente, la función `hstack()` realiza una concatenación a lo largo del segundo eje (axis=1), y los arreglos que toma como argumento deben coincidir en la forma a lo largo de este eje. Los arreglos unidimensionales pueden ser de cualquier longitud

```
#dimensiones mixtas
a = np.array([[1,2], 
              [3,4]])

b = np.array([[5, 6, 7],
              [8, 9, 10]])

print(a.shape, b.shape)
np.hstack([a, b])
```

```
#arreglos unidimensionales
a = np.array([1,2,3])

b = np.array([4,5,6,7])

print(a.shape, b.shape)
np.hstack([a, b])
```

Lo opuesto a la concatenación es la división, que se implementa mediante las funciones `split()`, `hsplit()` y `vsplit()`. Para cada uno de estas, podemos pasar una lista de índices que dan los puntos de división:

```
a = np.array([1, 2, 3, 99, 99, 3, 2, 1])

a1, a2, a3 = np.split(a, [3, 5])

print(a1, a2, a3)
```

Note que dividir en N puntos conduce a N + 1 subarreglos. Las funciones `hsplit` y `vsplit` funcionan de manera similar:

```
a = np.arange(16).reshape((4, 4))
a
```

```
superior, inferior = np.vsplit(a, 2)

print(superior)
print()
print(inferior)
```

```
izq, der = np.hsplit(a, 2)

print(izq)
print("")
print(der)
```

```
#para arreglos de dimension superior 
x = np.arange(8).reshape(2, 2, 2)
print(x.shape, x.ndim)
x
```

```
sup, inf = np.vsplit(x, 2)

print(sup)
print("")
print(inf)
```

```
izq, der = np.hsplit(x, 2)

print(izq)
print("")
print(der)
```

# 1.2 Indexación sofisticada

Anteriormente vimos cómo acceder y modificar porciones de arreglos usando índices simples (por ejemplo, `arr[0]`), segmentos (por ejemplo, `arr[: 5]`) y máscaras booleanas (por ejemplo, `arr[arr> 0]`). Veremos ahora otro estilo de indexación de arreglos, conocido como *indexación sofisticada*, la cual nos permite acceder y modificar muy rápidamente subconjuntos complicados de los valores de un arreglo.

La indexación sofisticada es conceptualmente simple: significa pasar una lista de índices en lugar de un entero, para acceder a múltiples elementos del arreglo a la vez. Veamos un ejemplo:

```
a = np.arange(10,20)
a
```

```
[a[1], a[3], a[7]]
```

```
# creamos lista con los indices
indices = [1, 3, 7]

# accedemos a los valores del arreglo
a[indices]
```

Con el indexado sofisticado, la forma del resultado refleja la forma del arreglo de índices más que la forma del arreglo que se está indexando:

```
a = np.arange(10)
a
```

```
# creamos arreglo de indices con forma (2,2)
ind = np.array([[3, 7],
                [4, 5]]) 

# accedemos a los valores del arreglo
a[ind]
```

El indexado sofisticado funciona también en múltiples dimensiones. Veámoslo en el siguiente ejemplo:

```
a = np.arange(12).reshape((3, 4))
a
```

Como en el indexado estándar, el primer índice hace referencia a las filas y el segundo a las columnas:

```
fila = [0, 1, 2]
col = [2, 1, 3]

a[fila, col]
```

Los valores el en arreglo corresponden a `a[0,2]`, `a[1,1]` y `a[2,3]` respectivamente. Podemos además combinar el indexado sofisticado con las otras técnicas de indexación que hemos visto para tener expresiones más complejas:

```
a
```

```
# combinando indexado sofisticado e índices simples
ind = [2, 0, 1]

a[2,ind]
```

```
# combinando indexado sofisticado con segmentacion
a[1:, ind]
```

```
# combinando indexado sofisticado con enmascaramiento
mas = np.array([1, 0, 1, 0], dtype=bool)
mas
```

```
a
```

```
a[:,mas]
```

# 1.3 Generación de números aleatorios

Una parte importante de cualquier simulación es la capacidad de generar números aleatorios. Para este propósito, NumPy proporciona varias rutinas en el submódulo `random`. Este utiliza un algoritmo particular, llamado Mersenne Twister, para generar números pseudoaleatorios.

Primero, necesitamos definir una semilla (seed) que haga predecibles los números aleatorios. Cuando se restablece el valor, los mismos números aparecerán cada vez. Si no asignamos la semilla, NumPy selecciona automáticamente un valor de semilla aleatorio basado en el dispositivo generador de números aleatorios del sistema o en el reloj.

Veamos a continuación las diferentes funciones para la generación de números aleatorios con NumPy

### Arreglo de valores aleatorios de punto flotante

Se puede generar un arreglo de valores aleatorios de punto flotante con la función `rand()` de NumPy. Esta crea un arreglo de una forma dada con valores aleatorias de una distribución uniforme sobre el intervalo ``[0, 1) ``.

> 
    np.random.rand(<forma>)


```
# creando un arreglo de la forma (1,)
np.random.rand(1)
```

```
# creando un arreglo de la forma (5,)
np.random.rand(5)
```

```
# creando un arreglo de la forma (2,2)
np.random.rand(2,2)
```

La función `np.random.random()` es similar a `np.random.rand()`, solo que esta toma explícitamente una tupla cuando se quieren crear arreglos con una forma del tipo `(a,b)` (no en el caso de formas del tipo `(a,)`)

```
# creando un arreglo de la forma (1,)
np.random.random(1)
```

```
# creando un arreglo de la forma (2,2)
np.random.random((2,2))
```

### Arreglo de valores aleatorios de tipo entero

Un arreglo de valores aleatorios de tipo entero se puede generar utilizando la función `randint()` de NumPy. 

Esta función toma tres argumentos: el extremo inferior, el extremo superior y el número de valores enteros a generar o la forma del arreglo. Los enteros aleatorios se extraerán de una distribución uniforme sobre el intervalo `[inferior, superior)`

> 
    np.random.randint(<inf>, <sup>, <forma>)


```
# creando un arreglo de la forma (1,) con valores en el intervalo [5,10)
np.random.randint(5,10)
```

```
# creando un arreglo de la forma (5,) con valores en el intervalo [5,10)
np.random.randint(5,10,5)
```

```
# creando un arreglo de la forma (2,2) con valores en el intervalo [5,10)
np.random.randint(5,10,(2,2))
```

Si no se especifica el límite superior, el intervalo a usar será `[0, inf)`

```
# creando un arreglo de la forma (1,) con valores en el intervalo [0,10)
np.random.randint(10)
```

```
# creando un arreglo de la forma (10,) con valores en el intervalo [0,10)
np.random.randint(10, size=10)
```

```
# creando un arreglo de la forma (2,2) con valores en el intervalo [0,10)
np.random.randint(10, size=(2,2))
```

### Sembrar las semilla del generador de números aleatorios

La función `np.seed()` se puede usar para sembrar la semilla del generador de NumPy de números pseudoaleatorios, tomando un entero como valor de la semilla:

```
# fijando la semilla
np.random.seed(4)

# creando el arreglo
np.random.rand(3)
```

```
# creando el arreglo
np.random.rand(3)
```

Otra método útil es ``random.choice()``, con el cual podemos seleccionar elementos de un arreglo de forma aleatoria:

```
a = np.arange(10)
a
```

```
np.random.choice(a)
```

Podemos pasar la forma del arreglo de salida:

```
np.random.choice(a, (2,2)) 
```

en incluso la probabilidad con la que queremeos que aparezca algún elemento en particular

```
np.random.choice(3, (2,2), [0,0.2,0.8])
```

## 2. Matplotlib

Matplotlib es una herramienta para visualización de datos en Python, construida sobre arreglos de NumPy. Usaremos el submódulo `pyplot` de la librería para realizar nuestros gráficos:

```
import matplotlib.pyplot as plt
```

Para realizar un gráfico de línea básico de $x$ vs $y$ utilizamos la función `plt.plot`

```
x = np.linspace(0,2*np.pi,100)
y = np.sin(x)

plt.plot(x,y);
```

Podemos agregar al gráfico un título y etiquetas para los ejes $x$ y $y$, con los métodos `plt.title`, `plt.xlabel` y `plt.xylabel`, respectivamente:

```
plt.plot(x, y)

# titulo de la figura
plt.title("función seno")

# etiquetas de los ejes
plt.xlabel("$x$")
plt.ylabel("$y(x)$");
```

Ahora, para definir los límites de los ejes tenemos dos opciones. La primera es utilizando los métodos `plt.xlim` y `plt.ylim`:

```
plt.plot(x, y)

# titulo de la figura
plt.title("función seno")

# etiquetas de los ejes
plt.xlabel("$x$")
plt.ylabel("$y(x)$");

# limites de los ejes
plt.xlim(1,5)
plt.ylim(-1.5,1.5)
```

O utilizando el método `plt.axis`, con el cual podemos especificar los límites de los ejes pasando una lista que contiene como elementos los límites de la siguiente manera:
>
    plt.axis([xmin, xmax, ymin, ymax])

```
plt.plot(x, y)

# titulo de la figura
plt.title("función seno")

# etiquetas de los ejes
plt.xlabel("$x$")
plt.ylabel("$y(x)$")

# limites de los ejes
plt.axis([1,5,-1.5,1.5]);
```

Con las palabras clave `tight` y `equal`, podemos definir los ejes de manera "ajustada" en el primer caso, y ejes donde una unidad en "$x$" sea igual a una unidad en "$y$" en el segundo caso:

```
plt.plot(x, y)

# titulo de la figura
plt.title("función seno")

# etiquetas de los ejes
plt.xlabel("$x$")
plt.ylabel("$y(x)$")

# limites de los ejes
plt.axis("equal");
#plt.axis("tight");
```

Para agregar una leyenda de la función, podemos pasar una lista con la(s) leyenda(s) a la función `plt.legend`:

```
plt.plot(x, y)

# titulo de la figura
plt.title("función seno")

# etiquetas de los ejes
plt.xlabel("$x$")
plt.ylabel("$y(x)$")

# limites de los ejes
plt.axis("equal");
#plt.axis("tight");

# leyenda de la linea
plt.legend(["Etiqueta de la gráfica"]);
```

O utilizando el argumento por palabra clave ``label`` dentro de la función ``plt.plot()``

```
plt.plot(x, y, label="Etiqueta de la gráfica")

# titulo de la figura
plt.title("función seno")

# etiquetas de los ejes
plt.xlabel("$x$")
plt.ylabel("$y(x)$")

# limites de los ejes
plt.axis("equal");
#plt.axis("tight");

# leyenda de la linea
plt.legend();
```

Nótese que debemos incluir ``plt.legend()`` para que se genere la leyenda. Además, podemos mostrar el título, los ejes y la leyenda, en un formato matemático encerrando la expresión entre los símbolos peso `"$"`

Podemos además especificar, mediante el argumento por palabra clave `loc`, la posición de la leyenda dentro del gráfico. Por defecto, se utiliza la "mejor" posición (best). Las opciones disponibles son:
>
    POSICIÓN        DESCRIPCIÓN
    best ---------- mejor
    upper right --- sup der
    upper left ---- sup izq
    lower left ---- inf izq 
    lower right --- inf der
    right --------- der 
    center left --- centro izq
    center right -- cento der
    lower center -- inf central
    upper center -- sup central
    center -------- central

```
plt.plot(x, y)

# titulo de la figura
plt.title("función seno")

# etiquetas de los ejes
plt.xlabel("$x$")
plt.ylabel("$y(x)$")

# limites de los ejes
plt.axis("equal");
#plt.axis("tight");

# leyenda de la linea
plt.legend(["Etiqueta de la gráfica"], loc="lower center");
```

### Graficando múltiples líneas

Si queremos mostrar múltiples líneas en la misma figura, simplemente tenemos que usar la función `plt.plot` repetidamente: 

```
x = np.linspace(-3*np.pi/2,3*np.pi/2)

plt.plot(x, np.sin(x), label="$sin(x)$")
plt.plot(x, np.cos(x), label="$cos(x)$")
plt.legend();
```

### Creando figuras y ejes

En la manera en que hemos venido creando nuestros gráficos (conocida como interfaz tipo MATLAB), la figura y los ejes se crean automáticamente: cuando utilizamos la función `plt.plot` se generan una figura y un conjunto de ejes de forma implícita

```
plt.plot(x, np.sin(x))
plt.xlabel("$x$")
plt.ylabel("$sin(x)$")
plt.axis([0,2*np.pi,-1,1]);
```

En el otro estilo de programación orientado a objetos, referenciamos la figura y los ejes mediante las variables `fig` y `ax` respectivamente, y serán métodos y atributos de estas variables los que nos permitan personalizar la visualización. Una primera opción para crear la figura y los ejes es utilizar las funciones `plt.figure` y `plt.axes`, que nos devuelven una referencia de la figura y conjunto de ejes utilizados.

```
# se crean la figura y los ejes
fig = plt.figure()
ax = plt.axes()
```

Podemos personalizar el estilo de nuestra figura mediante la directiva `plt.style`. Mediante el atributo `plt.style.available` podemos acceder a los diferentes estilos disponibles:

```
plt.style.available
```

Podemos elegir alguno de estos estilos utilizando el método `plt.style.use`:

```
plt.style.use("seaborn")
```

```
fig = plt.figure()
ax = plt.axes()
```

Una vez tenemos la referencia al conjunto de ejes `ax` podemos crear una gráfica ejecutando el método `plot` sobre `ax`:

```
# se crean la figura y los ejes
fig = plt.figure(figsize=(10,6))
ax = plt.axes()

# se realiza la grafica mediante la referencia al eje
ax.plot(x, np.sin(x));
```

Si bien la mayoría de las funciones en `plt` se traducen directamente a métodos `ax` (como `plt.plot()` → `ax.plot()`, `plt.legend()` → `ax.legend()`, etc.), este no es el caso para todos los comandos. En particular, las funciones para establecer límites, etiquetas y títulos se modifican ligeramente. Para realizar la transición entre las funciones de estilo MATLAB y los métodos orientados a objetos, realizamos los siguientes cambios:

* `plt.xlabel()` → `ax.set_xlabel()`
* `plt.ylabel()` → `ax.set_ylabel()`
* `plt.xlim()`   → `ax.set_xlim()`
* `plt.ylim()`   → `ax.set_ylim()`
* `plt.title()`  → `ax.set_title()`

La mayor parte de los métodos que nos permiten fijar un valor a un atributo comienza por "`set_`". Estos métodos suelen tener un método equivalente que comienza habitualmente por "`get_`" que sirve para leer el valor de dicho atributo.

Puede consultar todos los atributos y métodos de la clase `axes` en este [link](https://matplotlib.org/3.1.1/api/axes_api.html#array)

```
# se crean la figura y los ejes
fig = plt.figure(figsize=(5,3))
ax = plt.axes()

# graficando con la funcion ax.plot
ax.plot(x, np.sin(x), label="$sin(x)$")

# titulo de la figura
ax.set_title("Función Seno")

# etiquetas de los ejes
ax.set_xlabel("$x$")
ax.set_ylabel("$sin(x)$")

# limites de los ejes
ax.axis([0,2*np.pi,-1,1])

# leyenda de la funcion
ax.legend();
```

En la interfaz orientada a objetos, en lugar de llamar estas funciones individualmente, es más conveniente usar el método `ax.set()` para establecer todas estas propiedades a la vez:

```
fig = plt.figure()
ax = plt.axes()

ax.plot(x, np.sin(x), label="$sin(x)$")
ax.legend()

ax.set(xlim=(0,2*np.pi),
       ylim=(-1,1),
       xlabel="$x$",
       ylabel="$sen(x)$",
       title="Función Seno");
```

## Creando subgráficos

Para la generación de la figura y de uno (o más) conjuntos de ejes podemos recurrir a la función `plt.subplots`, la cual crea estos objetos y devuelve una referencia a la figura y al conjunto de ejes (si solo se ha creado uno) o a una lista de referencias a los conjuntos de ejes (si se han creado dos o más):

```
fig, ax = plt.subplots()

ax.plot(x, np.sin(x))
ax.set(xlabel="x",
       ylabel="sin(x)");
```

La función `plt.subplots` crea por defecto una figura y un solo eje. 

```
fig, ax = plt.subplots()

print(f"longitud de ax: {len(ax.plot(x, np.sin(x)))}");
```

Para tener dos gráficos en el mismo eje utilizamos la función `ax.plot` reiteradamente:

```
fig, ax = plt.subplots(figsize=(7,5))

ax.plot(x, np.sin(x), "ro", label="sin(x)")
ax.plot(x, np.cos(x), "g^", label="cos(x)")

ax.set(xlabel="x",
       title="Funciones Seno y Coseno")

ax.legend();
```

Si queremos especificar el número de ejes a usar, debemos hacerlo explícitamente en la función `plt.subplots`. Al igual que vimos en la sesión anterior con la función `plt.subplot`, la función `plt.subplots` toma como los dos primeros argumentos el número de filas y columnas a utilizar, solo que en este caso la referencia a las figuras se hace mediante `ax`:

```
fig, ax = plt.subplots(2, figsize=(10,6))

ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x));

print(f"Forma de ax: {ax.shape}")
```

En este caso `ax` es una lista con dos elementes, los cuales hacen referencia a los dos conjuntos de ejes creados `ax[0]` y `ax[1]` respectivamente. Para personalizar la figura, hacemos uso de esta referencia:

```
fig, ax = plt.subplots(2, figsize=(10,7))

ax[0].plot(x, np.sin(x), label="sin(x)")
ax[1].plot(x, np.cos(x), label="cos(x)")

ax[0].set(xlim=(0,2*np.pi),
       ylim=(-1,1),
       xlabel="$x$",
       title="Función Seno")

ax[1].set(xlim=(0,2*np.pi),
       ylim=(-1,1),
       xlabel="$x$",
       title="Función Coseno")

ax[0].legend()
ax[1].legend();
```

Note que en este caso, el título del segundo eje se está solapando con la etiqueta del eje x del primer eje. Para tener cada atributo de la figura en su lugar, podemos utilizar el argumento por palabra clave `constrained_layout` en la función `plt.subplots`:

```
fig, ax = plt.subplots(2, figsize=(10,6), constrained_layout=True)

ax[0].plot(x, np.sin(x), label="sin(x)")
ax[1].plot(x, np.cos(x), label="cos(x)")

ax[0].set(xlim=(0,2*np.pi),
       ylim=(-1,1),
       xlabel="$x$",
       title="Función Seno")

ax[1].set(xlim=(0,2*np.pi),
       ylim=(-1,1),
       xlabel="$x$",
       title="Función Coseno")

ax[0].legend()
ax[1].legend();
```

Otros argumentos por palabra clave útiles de la función `plt.subplots`, son `sharex` y `sharey`, que nos permiten tener mayor control sobre los ejes. Estos argumentos por palabra clave pueden tomar los siguientes valores: `True`, `False` o `{'none', 'all', 'row', 'col'}`:

```
fig, ax = plt.subplots(2, 2)

print(f"Forma de ax: {ax.shape}")
```

```
# Cada columna comparte el eje X
fig, ax = plt.subplots(2, 2, sharex="col")

ax[0,0].plot(x,np.sin(x))
ax[0,1].plot(x,np.cos(x))
ax[1,0].plot(x,np.sin(x))
ax[1,1].plot(x,np.cos(x));
```

```
# Cada fila comparte el eje Y
fig, ax = plt.subplots(2, 2, sharey='row')

ax[0,0].plot(x,np.sin(x))
ax[0,1].plot(x,np.cos(x))
ax[1,0].plot(x,np.sin(x))
ax[1,1].plot(x,np.cos(x));
```

```
# Todos los subgraficos comparten los ejes X y Y
fig, ax = plt.subplots(2, 2, sharex='all', sharey='all')
# equivalente a
#fig, ax = plt.subplots(2, 2, sharex=True, sharey=True)

ax[0,0].plot(x,np.sin(x))
ax[0,1].plot(x,np.cos(x))
ax[1,0].plot(x,np.sin(x))
ax[1,1].plot(x,np.cos(x));
```

## Gráficos de dispersión

Los gráficos de dispersión se utilizan para comparar los valores que toman dos variables distintas, una de las cuales se representa a lo largo del eje $x$ y la otra a lo largo del eje $y$. La gráfica resultante nos permite identificar visualmente la posible correlación entre las dos variables. Para realizar este tipo de gráficos podríamos utilizar la función `plt.plot` y remover las líneas. Sin embargo, Matplotlib ofrece una función mucho más potente `plt.scatter`. Esta función acepta como primeros dos argumentos los valores $x$ y $y$ de los puntos a mostrar. Hagamos un ejemplo sencillo: generemos valores aleatorios extraídos de una distribución normal para los valores de $x$ y de $y$:

```
import numpy as np
import matplotlib.pyplot as plt
plt.style.use("seaborn")

x = np.random.rand(20)
y = np.random.rand(20)

plt.scatter(x,y);
```

Existen casos en los que los datos se presentan en un formato que podemos acceder a variables particulares mediante una palabra clave, por ejemplo, cuando tenemos los datos almacenados en un diccionario:

```
dic = {"x":np.linspace(0,50)}
dic["y"] = dic['x'] + 10*np.random.randn(50)

plt.scatter("x","y", data=dic);
```

Este tipo de gráfica resulta más útil cuando estamos visualizando elementos de diferente clase y los mostramos en el gráfico con diferenes colores. El color de cada punto se controla mediante el argumento por palabra clave ``c``. Además, podemos cambiar el mapa de colores utilizando el argumento por palabra clave ``cmap`` y qué opaco es el color con el argumento por palabra clave ``alpha`` que va de 0 (transparente) a 1 (opaco).

*Puede consultar los mapas de colores disponibles en esta [dirección](https://matplotlib.org/3.1.0/tutorials/colors/colormaps.html)*

```
# generemos un arreglo para separar los datos en dos clases
# y asociarlas al color
color = (dic["y"] < 10)*1 + (dic["y"] > 10)*20
color
```

```
plt.scatter("x","y", c=color, alpha=0.5, data=dic, cmap="prism");
```

Podemos además modificar el tamaño de cada uno de los puntos mediante el argumento por palabra clave ``s``:

```
# el tamaño de los puntos aumenta a medida que x aumenta
s = np.arange(10,61)

plt.scatter("x","y", c=color, alpha=1, s=s, data=dic, cmap="prism");
```

```
# el tamaño de los puntos dependiendo del color
s = (dic["y"] < 10)*50 + (dic["y"] > 10)*150

plt.scatter("x","y", c=color, alpha=1, s=s, data=dic, cmap="prism");
```

Vamos a trabajar ahora con datos sencillos pero reales: el conjunto de datos Iris que contiene el ancho y el largo de los pétalos y sépalos de 150 flores de tres especies de Iris (setosa, versicolor y virginica)

```
from sklearn.datasets import load_iris

# se importan los datos
iris = load_iris()

datos = iris.data
clase = iris.target_names
caracteristicas = iris.feature_names

print(f"Forma de los datos: {datos.shape}")
print()
print(f"Clases: {clase}")
print()
print(f"Características: {caracteristicas }")
```

Vamos a cruzar las variables "petal_length" y "petal_width", información que cargaremos en las variables $x$ y $y$. 

```
x = datos[:,2]        # petal length
y = datos[:,3]        # petal width
colores = iris.target # colores asociados a cada tipo de flor

plt.scatter(x, y, c=colores, cmap="prism")
plt.xlabel("petal length")
plt.ylabel("petal width");
```

Es apreciable la distancia que separa el cluster representado por los puntos rojos (flores "setosa") de los otros dos que se muestran más mezclados. Si quisiéramos que la leyenda distinguiera entre 'setosa', 'versicolor' y 'virginica', tendríamos que crear tres gráficos de dispersión, uno para cada tipo de flor, y dar a cada uno una etiqueta diferente.
